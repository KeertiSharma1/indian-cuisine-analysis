```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("datasets/indian_food.csv")
df.head(100)
```

```python
df.info()
```

```python
# Replace -1 with NaN for consistency
df.replace(-1, pd.NA, inplace=True)
df.replace("-1", pd.NA, inplace=True)
df.head(100)
```

```python
# Fill the missing region value
df.loc[df['region'].isnull(), 'region'] = 'North'
print(df[df['name'] == 'Panjeeri'])
```

```python
# Handle numeric column
for col in ['prep_time', 'cook_time']:
    df[col] = pd.to_numeric(df[col], errors='coerce')
    df[col] = df.groupby("course")[col].transform(lambda x: x.fillna(x.median()))

df
```

```python
# Handle categorical columns
df['flavor_profile'].fillna("neutral", inplace=True)
df['state'].fillna("Unknown", inplace=True)
df['region'].fillna("Unknown", inplace=True)
df.head(100)
```

```python
# Create a new column for total time
df["total_time"] = df["prep_time"] + df["cook_time"]
df
```

```python
df.to_csv("datasets/indian_food_cleaned.csv", index=False)

print("✅ Cleaned data saved to datasets/indian_food_cleaned.csv")
```

```python
df['flavor_profile'].replace('-1', 'Unknown', inplace=True)
df['state'].replace('-1', 'Unknown', inplace=True)
df['region'].replace('-1', 'Unknown', inplace=True)
```

```python
df.isnull().sum()
```

```python
# Checking unique values in categorical columns
categorical_cols = ['flavor_profile', 'course', 'state', 'region']

for col in categorical_cols:
    print(f"\nUnique values in {col}:")
    print(df[col].unique())
    print(f"Total unique in {col}: {df[col].nunique()}")
```

```python
df['flavor_profile'].replace("Unknown", "neutral", inplace=True)
```

```python
# Top 5 states with the most dishes
top_states = df['state'].value_counts().head(5)

# Top 3 regions with the most unique states
state_region_counts = df.groupby('region')['state'].nunique().sort_values(ascending=False).head(3)

# Display results
print("Top 5 states with the most dishes:\n", top_states)
print("\nTop 3 regions with the most unique states:\n", state_region_counts)
```

```python
# Exclude "Unknown" from the state count
top_states = df[df['state'] != 'Unknown']['state'].value_counts().head(5)

# Display results again
print("Top 5 states with the most dishes (excluding 'Unknown'):\n", top_states)
```

```python
# Exclude "Unknown" from region count
top_regions = df[df['region'] != 'Unknown']['region'].value_counts().head(3)

# Display results again
print("Top 3 regions with the most unique states (excluding 'Unknown'):\n", top_regions)
```

```python
# Count of Vegetarian vs Non-Vegetarian
diet_counts = df['diet'].value_counts()

print("Diet Distribution:\n", diet_counts)
```

```python
# Grouping by state and diet type
state_diet_distribution = df.groupby(["state", "diet"]).size().unstack()

print("State-wise Diet Distribution:\n", state_diet_distribution)
```

```python
# Fill NaN with 0
state_diet_distribution = state_diet_distribution.fillna(0)
print(state_diet_distribution)
```

```python
state_diet_distribution.plot(kind='bar', stacked=True, figsize=(12, 6), colormap='viridis')

# Labels and title
plt.xlabel("State")
plt.ylabel("Count of Dishes")
plt.title("State-wise Diet Distribution (Vegetarian vs. Non-Vegetarian)")
plt.xticks(rotation=90)
plt.legend(title="Diet Type")

# Show plot
plt.show()
```

```python
# Plot the distribution of flavor profiles
plt.figure(figsize=(8, 5))
sns.countplot(x=df['flavor_profile'], palette="coolwarm")

# Labels and title
plt.xlabel("Flavor Profile")
plt.ylabel("Count of Dishes")
plt.title("Distribution of Flavor Profiles")

# Show the plot
plt.show()
```

```python
state_counts = df[df["state"] != "Unknown"]["state"].value_counts()

# Plot
plt.figure(figsize=(12, 6))
sns.barplot(x=state_counts.index, y=state_counts.values, palette="viridis")
plt.xticks(rotation=75)
plt.xlabel("State")
plt.ylabel("Number of Dishes")
plt.title("Number of Dishes by State")
plt.show()
```

```python
# Count dishes per region (excluding 'Unknown')
region_counts = df[df["region"] != "Unknown"]["region"].value_counts()

# Plot
plt.figure(figsize=(8, 8))
plt.pie(region_counts, labels=region_counts.index, autopct="%1.1f%%", colors=sns.color_palette("pastel"))
plt.title("Dish Distribution by Region")
plt.show()
```

```python
# Count dishes per state (excluding 'Unknown')
state_counts = df[df["state"] != "Unknown"]["state"].value_counts().head(5)

# Plot
plt.figure(figsize=(10, 5))
sns.barplot(x=state_counts.index, y=state_counts.values, palette="viridis")
plt.xlabel("State")
plt.ylabel("Number of Dishes")
plt.title("Top 5 States with Most Dishes")
plt.show()
```

```python
diet_counts = df["diet"].value_counts()

# Plot
plt.figure(figsize=(6, 6))
plt.pie(diet_counts, labels=diet_counts.index, autopct="%1.1f%%", colors=["green", "red"])
plt.title("Diet Distribution")
plt.show()
```

```python
# State-wise diet distribution (stacked bar chart)
state_diet = df.pivot_table(index="state", columns="diet", aggfunc="size", fill_value=0)

# Plot
state_diet.plot(kind="bar", stacked=True, figsize=(12, 6), colormap="viridis")
plt.title("State-wise Diet Distribution")
plt.xlabel("State")
plt.ylabel("Number of Dishes")
plt.xticks(rotation=90)
plt.legend(title="Diet")
plt.show()
```

```python
# Region-wise dish distribution
region_counts = df["region"].value_counts()

# Plot
plt.figure(figsize=(8, 8))
plt.pie(region_counts, labels=region_counts.index, autopct="%1.1f%%", colors=plt.cm.Paired.colors)
plt.title("Region-wise Dish Distribution")
plt.show()
```
